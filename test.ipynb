```python
from transformers import AutoTokenizer
import math

def estimate_code_chunks(file_path: str, model_name: str = "BAAI/bge-small-en-v1.5", chunk_size: int = 512, overlap: int = 50):
    # 1. Load the specific tokenizer you plan to use for embeddings
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    
    # 2. Read your target code file
    with open(file_path, 'r', encoding='utf-8') as f:
        code_text = f.read()
        
    # 3. Tokenize the raw code (bypassing the 512 truncation limit to get the full count)
    # add_special_tokens=False prevents adding [CLS]/[SEP] to the giant unchunked string
    tokens = tokenizer.encode(code_text, add_special_tokens=False)
    total_tokens = len(tokens)
    
    if total_tokens == 0:
        return 0, 0
        
    # 4. Calculate chunks using the sliding window mathematical formula
    stride = chunk_size - overlap
    
    if total_tokens <= chunk_size:
        num_chunks = 1
    else:
        # Subtract overlap from total, divide by stride, and round up
        num_chunks = math.ceil((total_tokens - overlap) / stride)
        
    print(f"--- RAG Chunking Estimation ---")
    print(f"Model: {model_name}")
    print(f"File: {file_path}")
    print(f"Total Tokens: {total_tokens}")
    print(f"Chunks Generated: {num_chunks} (Size: {chunk_size}, Overlap: {overlap})")
    
    return total_tokens, num_chunks

# Example usage for your hackathon test run:
tokens, chunks = estimate_code_chunks("micrograd/engine.py")
```
